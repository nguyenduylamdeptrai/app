```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

df = pd.read_csv('airbnb_cleaned_data.csv')

if 'name' in df.columns:
    df.drop('name', axis=1, inplace=True)

X = df.drop('price', axis=1)
Y = df['price']

categorical_cols = ['neighbourhood_group', 'neighbourhood', 'room_type']
# Chỉ lấy những cột chữ thực sự tồn tại trong dữ liệu hiện tại
cols_to_encode = [col for col in categorical_cols if col in X.columns]

# Thực hiện chuyển đổi
X_encoded = pd.get_dummies(X, columns=cols_to_encode, drop_first=True)

numerical_cols = ['minimum_nights', 'number_of_reviews', 'reviews_per_month',
                  'calculated_host_listings_count', 'availability_365', 'longitude', 'latitude']

cols_to_scale = [col for col in numerical_cols if col in X_encoded.columns]

scaler = StandardScaler()
X_encoded[cols_to_scale]= scaler.fit_transform(X_encoded[cols_to_scale])

print("Kích thước tập dữ liệu X sau khi biến đổi:", X_encoded.shape)
X_encoded.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X_encoded, Y, test_size = 0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# n_estimators=100: Tạo ra 100 cây quyết định
# n_jobs=-1: Dùng toàn bộ nhân CPU để chạy cho nhanh
rf_model = RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1)
rf_model.fit(X_train, Y_train)

# Dự đoán và Đánh giá
rf_pred = rf_model.predict(X_test)
rf_mae = mean_absolute_error(Y_test, rf_pred)
rf_rmse = np.sqrt(mean_squared_error(Y_test, rf_pred))
rf_r2 = r2_score(Y_test, rf_pred)

print("\n=== KẾT QUẢ RANDOM FOREST ===")
print(f"MAE: ${rf_mae:.2f}")
print(f"RMSE: ${rf_rmse:.2f}")
print(f"R-squared: {rf_r2:.4f}")
```

```python
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

xgb_model = XGBRegressor(n_estimators=100, learning_rate=0.1, random_state=42, n_jobs=-1)
xgb_model.fit(X_train, Y_train)

# Dự đoán và Đánh giá
xgb_pred = xgb_model.predict(X_test)
xgb_mae = mean_absolute_error(Y_test, xgb_pred)
xgb_rmse = np.sqrt(mean_squared_error(Y_test, xgb_pred))
xgb_r2 = r2_score(Y_test, xgb_pred)

print("\n=== KẾT QUẢ XGBOOST ===")
print(f"MAE: ${xgb_mae:.2f}")
print(f"RMSE: ${xgb_rmse:.2f}")
print(f"R-squared: {xgb_r2:.4f}")
```

```python
import joblib
from google.colab import files

# 1. Lưu mô hình tốt nhất (Giả sử là XGBoost)
# Nếu Random Forest của bạn tốt hơn, hãy thay xgb_model bằng rf_model
joblib.dump(xgb_model, 'airbnb_model.pkl')

# 2. Lưu công cụ chuẩn hóa (Scaler)
joblib.dump(scaler, 'airbnb_scaler.pkl')

print("Đã lưu thành công mô hình và scaler!")

# 3. Tải 2 file này về máy tính của bạn
files.download('airbnb_model.pkl')
files.download('airbnb_scaler.pkl')
```
